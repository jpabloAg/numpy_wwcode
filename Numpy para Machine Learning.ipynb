<img src="images/numpy_memory.png" style="width: 600px;">

# Listas vs Numpy array usando. un dataset de vinos 🍷
Ahora que sabemos como crear numpy arrays, podemos cargar un dataset en una lista de listas y empalmar con conocimientos exitentes de listas.

Vamos a utilizar un dataset de datos numericos de vinos, para determinar la calidad del vino teniendo en cuenta las siguientes variables:
* fixed acidity
* volatile acidity
* citric acid
* residual sugar
* chlorides
* free sulfur dioxide
* total sulfur dioxide
* density
* pH
* sulphates
* alcohol

````python
import csv
import numpy as np

# Leemos nuestro dataset como una lista de listas
f = open("files/winequality-white.csv", "r")
wine_quality = list(csv.reader(f, delimiter=';'))

# Eliminamos el encabezado
wine_quality = wine_quality[1:]

# convertimos todos los valores a float, ya que estan en str
lol_wines = []
for row in wine_quality:
    converted_row = []
    for item in row:
        converted_row.append(float(item))
    lol_wines.append(converted_row)

# Put in a numpy array
wines = np.array(lol_wines)
````

# Vectorización
Ya que notamos que seleccionar los datos con NumPy es más sencillo, vamos a aprender que hacer otras operaciones sobre los datos también es más eficiente, ya que evitamos ejecutar una acción cada vez que iteramos en el for loop,  y solamente efectuamos la operación una sola vez.

### Operaciones con vectorización:
Son operaciones que se realizan entre vectores, o ndarrays de la misma dimension y forma (shape)
* **Suma:** vector_a + vector_b
* **Resta:** vector_a - vector_b
* **Multiplicación:** vector_a * vector_b (no es el producto punto)
* **División:** vector_a / vector_b

*Para realizar operaciones, las columnas (o vectores) deben tener la misma dimensión, es decir, no puedo sumar un vector de 10 columnas con otro vector de 20 columnas, ya que se emparejan linea por línea.

La operación sólo se hace una vez!


Para ilustrar el ejemplo, vamos a sumar dos columnas del dataset
###  Con lista de listas


````python
total_quality_alcohol = []

alcohol = [-2]
ph = lol_wines[-4]

for row in lol_wines:
    quality = row[-1]
    alcohol = row[-2]
    quality_alcohol = quality + alcohol
    total_quality_alcohol.append(quality_alcohol)
    
total_quality_alcohol
````

### Con numpy:


````python
quality = wines[:,-1]
alcohol = wines[:,-2]

total_quality_alcohol = quality + alcohol
total_quality_alcohol
````

````python
# o en una sola línea, y (sabemos que es eficiente)
total_quality_alcohol = wines[:,-1] + wines[:,-2]
total_quality_alcohol
````

### Debo asegurarme que las dimensiones de los ndarrays sean iguales, de lo contrario, no se puede hacer vectorizacion.

````python
# esto fallará:
total_quality_alcohol = wines[:,-1] + wines[50:,-2]
````

## Reshaping
Es importante recalcar que los números "cuadren" al hacer reshaping, ya que no podemos perder datos, el reshaping debe contemplar la cantidad de datos, esto se puede averiguar multiplicando los elemtos de su forma (.shape):

## Por qué es importante el reshaping?
Algunas librerías, como scikit-learn, pueden requerir que una matriz unidimensional de variables de salida (y) tenga la forma de una matriz bidimensional con una columna y resultados para cada fila.

Algunos algoritmos, en Keras, requieren que la entrada se especifique como una matriz tridimensional compuesta de muestras, pasos de tiempo y características.

Es importante saber cómo remodelar los ndarrays NumPy para que los datos cumplan con las expectativas de ciertas librerías.

### De 2-D array a 1-D array

````python
reshaper = np.array([[1, 2, 3],
                     [4, 5, 6]])

print(reshaper.shape)

# fallara, ya que para pasar a la dimensión anterior se debe considerar 2*3 = 6
# Una sola dimensión de dos elementos, es decir, perderíamos 4 datos
reshaper.reshape(2)
````

````python
# Usando una sola dimensión de 6 elementos
reshaper.reshape(6)
````

### De 2-D array a 3-D array

````python
# no fallará
# Una sola dimensión de 6 elementos:
another_reshaper = np.array([[1, 2, 3, 0],
                     [4, 5, 6, 0]])
another_reshaper.reshape(2, 2, 2)
````

## Reshaping un 3-D array a 2-D array:

````python
tensor_3d_copy = tensor_3d.copy()
print(tensor_3d_copy.shape)
tensor_3d_copy
````

````python
tensor_3d_copy.reshape(4, 4)
````

## Diferencia entre métodos y funciones:
Cada ndarray tiene los metodos anteriormente mencionados, y numpy tiene funciones que podemos invocar enviando el ndarray como argumento:


````python
# Con funciones de numpy (recordemos que usamos el alias np para numpy al importarlo).
quality = wines[-1]
print(np.min(quality))
print(np.max(quality))
print(np.mean(quality))
print(np.median(quality))
````

## Ejercicio, traer las calidades de vino iguales o superiores a 9

## Índices booleanos
Podemos filtrar los datos, dependiendo de los criterios que necesitemos. Por ejemplo, si queremos traer cuántos vinos tienen una calidad igual o superior a 7:

````python
total_wines = wines.shape[0]
top_wines = wines[:, -1] > 7
percentage_top = (wines[top_wines].shape[0] / total_wines) * 100
percentage_top = round(percentage_top, 2)
percentage_top
````

### Operadores lógicos para índices booleanos
Se pueden utilizar operadores lógicos para verificar más de una condición,  a diferencia de python, numpy utiliza diferentes maneras de usar los operadores:
* El operador lógico not en numpy es ~
* El operador lógico and en numpy es &
* El operador lógico or en numpy es |

````python
mid_wines_upper = wines[:, -1] < 7
mid_wines_under = wines[:, -1] >= 5.0
# Para saber cuantos vinos en total.

percentage_mid = (wines[mid_wines_upper & mid_wines_under].shape[0] / total_wines) * 100
percentage_mid = round(percentage_mid, 2)
percentage_mid
````

````python
bad_wines = wines[:, -1] < 5
percentage_bad = (wines[bad_wines].shape[0] / total_wines) * 100
percentage_bad = round(percentage_bad, 2)
percentage_bad
````
